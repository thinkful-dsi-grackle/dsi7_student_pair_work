```python
import pandas as pd

df = pd.read_csv('https://tf-assets-prod.s3.amazonaws.com/tf-curric/data-science/web_events.csv')
df
```

```python
df['timestamp'] = pd.to_datetime(df['timestamp'],unit='ms')
df
```

```python
from datetime import datetime as dt
df['year'] = df['timestamp'].dt.year
df['month'] = df['timestamp'].dt.month
df['day'] = df['timestamp'].dt.day
df
```

```python
df['hour_round'] = df['timestamp'].dt.round('H')
df['hour'] = df['hour_round'].dt.hour
df
```

```python
i = range(0,24)
view = []
add = []
tran = []
for x in i:
    print('hour = ', x)
    print(df[df['hour'] == x]['event'].value_counts())
    view.append(df[df['hour'] == x]['event'].value_counts()[0])
    add.append(df[df['hour'] == x]['event'].value_counts()[1])
    tran.append(df[df['hour'] == x]['event'].value_counts()[2])
```

```python
import matplotlib.pyplot as plt
plt.plot(range(0,24), view,label = 'view')
plt.plot(range(0,24), add,label = 'add to cart')
plt.plot(range(0,24), tran,label = 'transaction')
plt.legend()
```

```python
df1 = pd.read_csv('https://tf-assets-prod.s3.amazonaws.com/tf-curric/data-science/life_expectancy.csv')
df1
```

```python
ids=['Country Name', 'Country Code', 'Indicator Name', 'Indicator Code']
melt_fields = set(df1.columns) - set(ids)
melted = pd.melt(df1,id_vars=ids,
                value_vars=melt_fields,
                var_name='Year',value_name='Value')
melted.head()
```

```python
#Imputation
melted['Value'] = melted['Value'].fillna(melted['Value'].mean())
melted.head()
```

```python
#interpolation
ids=['Country Name', 'Country Code', 'Indicator Name', 'Indicator Code']
melt_fields = set(df1.columns) - set(ids)
melted = pd.melt(df1,id_vars=ids,
                value_vars=melt_fields,
                var_name='Year',value_name='Value')
melted['Value'] = melted['Value'].fillna(method='ffill')
melted.head()
```

```python
#deletion
ids=['Country Name', 'Country Code', 'Indicator Name', 'Indicator Code']
melt_fields = set(df1.columns) - set(ids)
melted = pd.melt(df1,id_vars=ids,
                value_vars=melt_fields,
                var_name='Year',value_name='Value')
melted = melted.dropna()
melted.head()
```
