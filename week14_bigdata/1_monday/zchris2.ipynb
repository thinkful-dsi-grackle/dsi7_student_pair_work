How many counties have a unemployement rate above 10%?

```python
#Answer below:
len(census[census['Unemployment'] > 10].County.unique().compute())
```

Print the number of counties in each state.

```python
# Answer below:
census.groupby('State')['County'].nunique().compute()
```

Add a high child poverty boolean column. This column will be 1 if the rate of child poverty is over %20 and 0 otherwise.

```python
# Answer below:
census['HighChildPov'] = census['ChildPoverty'].map(lambda x: 1 if x>20 else 0)
census['HighChildPov'].compute()
```

What are the top 10 most populated counties?

```python
#answer below:
population = census.groupby('County').sum()['TotalPop'].compute()
population.sort_values(ascending=False).head(10)
```
