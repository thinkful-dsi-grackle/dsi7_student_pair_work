### Scrape the HTML from the URL below which contains questions and answers about databases.

```python
url = 'https://www.wisdomjobs.com/e-university/database-interview-questions.html'
```

```python
response = requests.get(url)

soup = BeautifulSoup(response.text)
soup
```

### Clean the acquired HTML, extracting just the text.

```python
quesans = soup.find_all('li', attrs={'class':'quesans'})
quesans = [q.text for q in quesans]
quesans
```

### Organize the text into a list of questions and answers.

The best way to do this is to use a regular expression to split the text wherever a question or an answer appears (they are labeled).

```python
questions = []
answers = []
for qa in quesans:
  qa_split = qa.split('Answer :')
  questions.append(qa_split[0])
  answers.append(qa_split[1])
```

```python
questions = [re.sub(r'Question \d*\. ', '', q).replace('\n', '') for q in questions]
questions
```

```python
answers = [a.replace('\n', ' ') for a in answers]
answers
```

```python
len(questions)
```

```python
len(answers)
```

```python
qa = []
for i in range(len(questions)):
  qa.append(questions[i])
  qa.append(answers[i])
qa
```

### Train a Chatterbot chatbot on the list of questions and answers.

```python
chatbot = ChatBot('SQLbot')
listtrainer = ListTrainer(chatbot)

listtrainer.train(qa)
```

### Write the user interface logic that allows a user to ask the chatbot questions about databases and have the chatbot return an answer. 

Include logic that checks to see if the answer returned is going to be a question and if so, return the next element in the list after that question.

```python
chatbot.get_response('Explain What Is Sqlite Transactions?')
```

```python

print("Hi, I'm SQLbot. Ask me a database question.")
print('Type EXIT to end the session.')

while True:
  user = input('User: ')

  if user.lower() != 'exit':
    response = chatbot.get_response(user).text
    if response.endswith('?'):
      print('SQLbot: ', qa[qa.index(response) + 1])
    else:
      print('SQLbot: ', response)
  else:
    print('Thank you')
    break
```
