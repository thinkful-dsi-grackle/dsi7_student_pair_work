# Identify 3 scenarios in your everyday life where you use probability.
  1. Going to the gym: will I get a machine? 
  2. Will I use my umbrella in AZ?
  3. Will my partner like the gift I get them?

# Using the formula you learned to calculate probabilities, approximate the probabilities of the events in each of those 3 scenarios.
### P(E) = Event Frequency / Total Observations
  1. P(getting a machine) = how often I've gotten a machine / total number of times I've tried
  
    P(getting a machine) = 80%  

  2. P(I'll use my umbrella) = how often it's rained in AZ / number of days I've observed
  
    P(I'll use my umbrella) = 3%

  3. P(my partner likes the gift) = how many times they've liked my gifts / how many gifts I've given
  
    P(my partner likes the gift) = 60%

#Think about what potential dependencies the events in each of your 3 scenarios have. Calculate the joint probabilities for the events and their dependencies.
###P(AB) = P(A|B) x P(B)

  1. Whether or not the gym is open is a dependency:
  
    A|B: Getting a machine given then gym is open (80%) B: gym is open (90%)
    
    P(the gym is open and you get a machine) = .8 * .9 = .72
    
    72% chance the gym will be open and you'll get a machine


  2. If I can fit an umbrella in my bag is a dependency:

    A|B: I'll need an umbrella given I can fit it in my bag (3%) B: I can fit it in my bag (50%)

    P(I'll need an umbrella and it will fit) = .03 * .5 = .015

    1.5% chance that I'll need an umbrella and will be able to fit it in my bag

  3. A|B: partner likes gift, given it arrives on time (60%) B: it arrives on time (20%)

    P(partner likes gift and it arrives on time) = .6 * .2 = .12

    12% chance that the gift arrives on time and my partner likes it

#Calculate the conditional probabilities for the events given their dependencies.
###P(A|B) = P(AB) / P(B)

1. AB: Getting a machine & gym is open (72%) B: Gym is open (90%)

    P(getting a machine given that the gym is open) = .72 / .9 = .8

2. AB: I use my umbrella & it fits in my bag (1.5%) B: it fits in my bag (50%)
    
    P(I use my umbrella given that it fits in my bag) = .015 / .5 = .03

3. AB: that the gift arrives on time and my partner likes it (12%) B: it arrives on time (20%)

    P(my partner likes the gift, given that it arrives on time) = .12 / .2 = .6
