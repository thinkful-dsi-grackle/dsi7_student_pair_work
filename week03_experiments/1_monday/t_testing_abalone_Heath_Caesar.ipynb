```python
import pandas as pd
import numpy as np 
from scipy import stats
import matplotlib.pyplot as plt


df = pd.read_csv('https://tf-assets-prod.s3.amazonaws.com/tf-curric/data-science/abalone_csv.csv')
```

```python
df
```

```python
abalone_male=df.loc[df['Sex'] == 'M']
abalone_female=df.loc[df['Sex'] == 'F']
print(stats.describe(abalone_male['Whole_weight']))
print(stats.describe(abalone_female['Whole_weight']))
print(stats.ttest_ind(abalone_male['Whole_weight'], abalone_female['Whole_weight']))
plt.hist(abalone_male['Whole_weight'])
plt.hist(abalone_female['Whole_weight'])
```

Formal Hypotheses: 
H0 - There is no significant difference between the whole weight of male abalone and female abalone

Ha - There is a significant difference between the whole weight of male abalone and female abalone

Given the results of the T-test, the T-statistic is ~ -3.23 and the p-value is ~ 0.001, so there is a 0.1% chance the that we can accept the null hypothesis, i.e. there is a 0.1% chance the difference in whole weight between male and female abalone is due to random chance.
