# Descriptive Statistics Assignment

```python
import pandas as pd

data_url = "https://docs.google.com/spreadsheets/d/1w1BfAGjEJk_Ywglo0nWKUk5gKmqEmfFb101IoXtj8D8/export?format=csv"
```

### Read the aggregated AirBnB listings of New York (using `data_url`) into a Pandas data frame.

```python
airbnb = pd.read_csv(data_url)
```

### Print the data type for each column.

```python
airbnb.dtypes
```

### Generate a basic set of descriptive statistics for the numeric fields in the data set.

```python
numeric_stats = airbnb.describe().T
numeric_stats
```

### Calculate the following for each numeric field in the data set and add them to the data frame containing the basic set of descriptive statistics.

- Variance
- Mean Absolute Deviation
- Range
- Interquartile Range

```python
numeric_stats['var'] = airbnb.var()
numeric_stats['mad'] = airbnb.mad()
airbnb_int = airbnb.select_dtypes(exclude=['object'])
numeric_stats['rng'] = airbnb_int.max() - airbnb_int.min()
numeric_stats['IQR'] = numeric_stats['75%'] - numeric_stats['25%']

numeric_stats
```

### Generate a basic set of descriptive statistics for the categorical fields in the data set.

```python
categorical = airbnb.describe(include='object').T
categorical
```

### Calculate the Pearson correlation coefficients between price, minimum_nights, reviews_per_month, calculated_host_listings_count and availability_365

```python
airbnb[['price', 'minimum_nights', 'reviews_per_month', 'calculated_host_listings_count', 'availability_365']].corr()
```

### Group the data by neighbourhood and calculate the mean and standard deviation of price for each neighbourhood.

```python
aggs = {'price':['mean', 'std']}
n_stats = airbnb.groupby(['neighbourhood']).agg(aggs).reset_index()
n_stats.head(10)
```

### Create a new minimum revenue index variable that combines the price, minimum nights, calculate host listings count into a single value for each dwelling. Then sort the data frame by that value to see which dwelling has the most financial strength. 



```python
airbnb['MRI'] = (airbnb['price']) + \
                (airbnb['minimum_nights']) + \
                (airbnb['calculated_host_listings_count'])

airbnb.sort_values('MRI', ascending=False).head(10).reset_index()
```

### Group the data by neighbourhood_group and calculate the average minimum revenue index for each neighbourhood_group.

```python
aggs2 = {'MRI':['mean']}
ng_stats = airbnb.groupby(['neighbourhood_group']).agg(aggs2).reset_index()
ng_stats.head()
```
