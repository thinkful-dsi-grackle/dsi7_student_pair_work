# Descriptive Statistics Assignment

```python
import pandas as pd

data_url = "http://data.insideairbnb.com/united-states/ny/new-york-city/2020-06-08/visualisations/listings.csv"
```

### Read the aggregated AirBnB listings of New York (using `data_url`) into a Pandas data frame.###

```python
df_airbnb = pd.read_csv(data_url)

df_airbnb.head()
```

### Print the data type for each column.###

```python
df_airbnb.dtypes
```

### Generate a basic set of descriptive statistics for the numeric fields in the data set.###

```python
numeric_stats = df_airbnb.describe().T

numeric_stats
```

### Calculate the following for each numeric field in the data set and add them to the data frame containing the basic set of descriptive statistics.###

- Variance
- Mean Absolute Deviation
- Range
- Interquartile Range

```python
numeric_stats.index
```

```python
numeric_stats['var'] = numeric_stats['std']**2
numeric_stats['mad'] = [df_airbnb[index].mad() for index in numeric_stats.index]
numeric_stats['range'] = numeric_stats['max'] - numeric_stats['min']
numeric_stats['iqr'] = numeric_stats['75%'] -numeric_stats['25%']
```

```python
numeric_stats
```

### Generate a basic set of descriptive statistics for the categorical fields in the data set.###

```python
categorical_stats = df_airbnb.describe(include='object').T

categorical_stats
```

### Calculate the Pearson correlation coefficients between price, minimum_nights, reviews_per_month, calculated_host_listings_count and availability_365###

```python
df_airbnb[['price', 'minimum_nights', 'reviews_per_month', 'calculated_host_listings_count', 'availability_365']].corr()
```

### Group the data by neighbourhood and calculate the mean and standard deviation of revenue for each neighbourhood. ###

```python
aggs = {'price': ['mean', 'std']}
neighbourhood_stats = df_airbnb.groupby(['neighbourhood']).agg(aggs).reset_index()
neighbourhood_stats.head(10)
```

### Create a new mimimum revenue index variable that combines the price, minimum nights, calculate host listings count into a single value for each dwelling. Then sort the data frame by that value to see which dwelling has the most financial strength. ###



```python
df_airbnb['min_riv'] = df_airbnb['price'] + df_airbnb['minimum_nights'] + df_airbnb['calculated_host_listings_count']

df_airbnb.sort_values('min_riv', ascending=False).head(10)
```

### Group the data by neighbourhood_group and calculate the average minimum revenue index for each neighbourhood_group. ###

```python
aggs = {'price':['mean'], 'minimum_nights': ['mean'], 'calculated_host_listings_count':['mean']}
neighbourhood_group_data = df_airbnb.groupby(['neighbourhood_group']).agg(aggs).reset_index()
neighbourhood_group_data['min_riv'] = neighbourhood_group_data[('price', 'mean')] + \
                                      neighbourhood_group_data[('minimum_nights', 'mean')] + \
                                      neighbourhood_group_data[('calculated_host_listings_count', 'mean')]

neighbourhood_group_data.head()
```

# Probability Theory Assignment

1. Identify 3 scenarios in your everyday life where you use probability.


a. What is the chance of having pizza for dinner today?

b. Given it is raining today, what is the chance of a morning run?

c. Going to the supermarket, what is the chance of buying both milk and cereal?

2. Using the formula you learned to calculate probabilities, approximate the probabilities of the events in each of those 3 scenarios.

a. P(eat pizza) = 20%

b. P(morning_run) = 60%

c. P(milk) = 80%   P(cereal) = 30%

3. Think about what potential dependencies the events in each of your 3 scenarios have. Calculate the joint probabilities for the events and their dependencies.

a. Having coke with pizza.
P(pizza & coke) = P(pizza) * P(coke) = 20% * 20% = 4%

b. Add a few post-running push-ups
P(run & pushup) = P(run) * P(pushup) = 60% * 10% = 6%

c. buying both items.
P(milk & cereal) = P(milk | cereal) * P(cereal) = 90% * 30% = 27%

4. Calculate the conditional probabilities for the events given their dependencies.

a. Had pizza last night.
P(pizza | pizza yesterday) = P(pizza & pizza yesterday) / P(pizza yesterday) = 2% / 20% = 10%

b. If it is raining outside.
P(run | rain) = P(run & rain) / P(rain) = 1% / 20% = 5%

c. Buying milk given buying cereal.
P(milk | cereal) = P(milk & cereal) / P(cereal) = 27% / 30% = 90%
