# Descriptive Statistics Assignment

```python
import pandas as pd

data_url = "http://data.insideairbnb.com/united-states/ny/new-york-city/2020-06-08/visualisations/listings.csv"
```

### Read the aggregated AirBnB listings of New York (using `data_url`) into a Pandas data frame.

```python
df = pd.read_csv(data_url)
df
```

### Print the data type for each column.

```python
df.dtypes
```

### Generate a basic set of descriptive statistics for the numeric fields in the data set.

```python
stats = df.describe()
dfNum = df.select_dtypes("number")
stats
```

### Calculate the following for each numeric field in the data set and add them to the data frame containing the basic set of descriptive statistics.

- Variance
- Mean Absolute Deviation
- Range
- Interquartile Range

```python
variance = pd.Series(df.var(), name="variance")
mad = pd.Series(df.mad(), name="mean absolute deviation")
range = pd.Series(dfNum.max()-dfNum.min(), name="range")
iqr = pd.Series(df.quantile(q=0.75)-df.quantile(q=0.25), name="iqr")
stats = stats.append(variance)
stats = stats.append(mad)
stats = stats.append(range)
stats = stats.append(iqr)
stats

```

### Group the data by neighbourhood and calculate the mean and standard deviation of revenue for each neighbourhood.

```python
aggs = {'price': ['mean', 'std']}
neighborhoods = df.groupby(['neighbourhood']).agg(aggs)
neighborhoods
```

### Create a new mimimum revenue index variable that combines the price, minimum nights, calculate host listings count into a single value for each dwelling. Then sort the data frame by that value to see which dwelling has the most financial strength. 



```python
df['minimum_revenue_index'] = (df['price'] * df['minimum_nights'] * df['calculated_host_listings_count'])
df.sort_values('minimum_revenue_index', ascending=False).head(10).reset_index().T
```

### Group the data by neighbourhood_group and calculate the average minimum revenue index for each neighbourhood_group.

**Example 2**

The chance of an accident happening at a Mal-Wart store on a regular day is 1%. On Black Friday, the chance is 20%. An accident report was handed in without a date on it, what is the probability the accident happened on Black Friday?

P(Black Friday) = 1/365 - Chance that the day is Black Friday

P(Regular Day) = 364/365 - Chance that the day is a regular day

P(Accident|Regular Day) = 1/100 - Chance of accidents on a regular day

P(Accident|Black Friday) = 1/2 - Chance of accidents on Black Friday

P(Accident) = P(Black Friday) * P(Accident|Black Friday) + P(Regular Day) * P(Accident|Regular Day) - Chance of an accident on any given day

P(Accident) = 1/7300 + 364/36500 - Chance of an accident on any given day

P(Accident) = 369/36500 - Chance of an accident on any given day

P(Black Friday|Accident) = P(Accident|Black Friday) * P(Black Friday) / P(Accident) - Chance that any given accident happened on Black Friday

P(Accident|Black Friday) * P(Black Friday) = 1/5 * 1/365 = 1/1825 - Likelihood * Prior

P(Black Friday|Accident) = (1/1825)/(369/36500) = 20/369 = 5.542% - Chance that any given accident happened on Black Friday  


If there is an accident in the store, there is a 5.542% chance it is Black Friday, which means there is a 94.5% chance the accident did not happen on Black Friday.

Since we know 92 batteries will get through the machine and two of those are SD and other 90 are good we want to get proportion of good batteries divided by all the batteries that passes. Thus 90/92 = .978, so there is 97.8% chance that the battery that will be shipped are good.
